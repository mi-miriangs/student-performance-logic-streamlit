
# Student Performance Dataset - Análise Bronze + PyCaret

Este notebook faz uma exploração simples (**nível Bronze**) do dataset *Students Performance in Exams*,
aplicando conceitos de **Matemática para Computação, Conjuntos, Lógica Proposicional e Estatística**,
e finalizando com um exemplo de **PyCaret** para classificação.

---


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# carregar dataset (ajuste o caminho se necessário)
df = pd.read_csv("data/StudentsPerformance.csv")

df.head()
```

```python

# Visão geral do dataset
df.info()
```

```python

# Estatísticas descritivas apenas para colunas numéricas
df.select_dtypes(include="number").describe()
```

```python

# Histograma das notas de matemática
sns.histplot(df["math score"], bins=15, kde=True)
plt.title("Distribuição das notas de Matemática")
plt.show()

# Boxplot de matemática por gênero
sns.boxplot(x="gender", y="math score", data=df)
plt.title("Boxplot: Math Score por Gênero")
plt.show()
```

```python

# Conjuntos de aprovados (nota >= 60)
aprov_math = set(df.index[df["math score"] >= 60])
aprov_read = set(df.index[df["reading score"] >= 60])

print("Aprovados em ambos (interseção):", len(aprov_math & aprov_read))
print("Aprovados em pelo menos uma (união):", len(aprov_math | aprov_read))
print("Só em matemática:", len(aprov_math - aprov_read))
```

```python

# Coluna 'aprovado': se math >= 60 E reading >= 60
df["aprovado"] = (df["math score"] >= 60) & (df["reading score"] >= 60)

df[["math score", "reading score", "aprovado"]].head()
```

```python

# Probabilidade geral de aprovação
prob_aprov_geral = df["aprovado"].mean()

# Probabilidade por gênero
prob_aprov_por_genero = df.groupby("gender")["aprovado"].mean()

print("Probabilidade geral de aprovação:", prob_aprov_geral)
print("Probabilidade de aprovação por gênero:")
print(prob_aprov_por_genero)
```

```python

# Exemplo com PyCaret
from pycaret.classification import setup, compare_models

# Configuração do PyCaret (coluna 'aprovado' como alvo)
clf = setup(
    data=df,
    target="aprovado",
    session_id=123,
    normalize=True,
    verbose=False
)

# Comparação de modelos básicos
best_model = compare_models()
best_model
```
